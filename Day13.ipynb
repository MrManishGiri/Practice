_Let's see if you can write a function to help clean US zip code data. Given a string, it should return whether or not that string represents a valid zip code. For our purposes, a valid zip code is any string consisting of exactly 5 digits._

```python
def is_valid_zip(zip_code):
    """Returns whether the input string is a valid (5 digit) zip code
    """
    if len(zip_code)== 5 and zip_code.isdigit():
        return True
    return False
is_valid_zip("12345")
```

# 2.
**_A researcher has gathered thousands of news articles. But she wants to focus her attention on articles including a specific word. Complete the function below to help her filter her list of articles._**

_Your function should meet the following criteria:_

* Do not include documents where the keyword string shows up only as a part of a larger word. For example, if she were looking for the keyword “closed”, you would not include the string “enclosed.”

* She does not want you to distinguish upper case from lower case letters. So the phrase “Closed the case.” would be included when the keyword is “closed”

* Do not let periods or commas affect what is matched. “It is closed.” would be included when the keyword is “closed”. But you can assume there are no other types of punctuation.

```python
import re
def word_Search(doc_list, keyword):
    """
    Takes a list of documents (each document is a string) and a keyword.
    Returns list of the index values into the original list for all documents
    containing the keyword.

    Example:
    doc_list = ["The Learn Python Challange Casino.", "They bought a car", "Casionville"]
    >>> word_search(doc_list, 'casino')
    >>> [0]
    """
    #list to store the indices of documents containing the keyword
    indices = []

    # Normalize the keyword to lowercase for case_insensitive matching
    keyword = keyword.lower()

    # Regular expression to match the keyword as a whole word, ignoring punctuation
    keyword_pattern = r'\b' + re.escape(keyword) + r'\b'

    for i, doc in enumerate(doc_list):
        #Normalize the document to lowercase
        normalized_doc = doc.lower()

        # Remove periods and commas from the document
        normalize_doc = normalized_doc.replace('.','').replace(',','')

        #check if the keyword is present as a whole word in the document
        if re.search(keyword_pattern, normalized_doc):
            indices.append(i)

    return indices

doc_list = ["The Learn Python Challenge Casino.", "They bought a car", "Casinoville"]
print(word_Search(doc_list, 'casino'))  # Should return [0]
print(word_Search(doc_list, 'car'))     # Should return [1]
print(word_Search(doc_list, 'Python'))  # Should return [0]
```

# 3.
_Now the researcher wants to supply multiple keywords to search for. Complete the function below to help her._

_(You're encouraged to use the word_search function you just wrote when implementing this function. Reusing code in this way makes your programs more robust and readable - and it saves typing!)_

```python
import re

def word_search(doc_list, keyword):
    """
    Takes a list of documents (each document is a string) and a keyword. 
    Returns a list of the index values into the original list for all documents 
    containing the keyword.
    """
    
    indices = []
    keyword = keyword.lower()
    keyword_pattern = r'\b' + re.escape(keyword) + r'\b'
    
    for i, doc in enumerate(doc_list):
        normalized_doc = doc.lower().replace('.', '').replace(',', '')
        if re.search(keyword_pattern, normalized_doc):
            indices.append(i)
    
    return indices

def multi_word_search(doc_list, keywords):
    """
    Takes list of documents (each document is a string) and a list of keywords.  
    Returns a dictionary where each key is a keyword, and the value is a list of indices
    (from doc_list) of the documents containing that keyword

    >>> doc_list = ["The Learn Python Challenge Casino.", "They bought a car and a casino", "Casinoville"]
    >>> keywords = ['casino', 'they']
    >>> multi_word_search(doc_list, keywords)
    {'casino': [0, 1], 'they': [1]}
    """
    result = {}
    
    for keyword in keywords:
        result[keyword] = word_search(doc_list, keyword)
        
    return result
```
